**Dataset**: funalab/CoND · SH-SY5Y human neuroblastoma · phase contrast · 176 images  
**Task**: Binary classification — differentiated (1) vs undifferentiated (0)  
**Config**: `config/config_CoND.yaml`

|--------|-----------|------|

## Module 1 — Preprocessing

**What changed and why:**

1. **Folder structure** — CoND organises images into `differentiated/` and `undifferentiated/` subfolders. The prototype read a flat directory (`data/raw/`) and parsed the timepoint from a filename regex. Here we walk two subfolders and assign the label directly from the folder name.

2. **Label assignment** — `differentiated → 1`, `undifferentiated → 0`. No regex needed.

3. **Resize target** — 200×200 (CoND native size) instead of 512×512. Resizing is mostly a no-op since images are already 200×200, but kept for robustness.

4. **Output** — instead of a patch manifest, this module produces an `image_manifest` DataFrame (one row per image, not per patch).

```python
def load_image(path):
    """Load PNG/TIFF as grayscale float32 in [0, 1]."""
    img = cv2.imread(str(path), cv2.IMREAD_GRAYSCALE)
    if img is None:
        raise ValueError(f"Cannot load: {path}")
    img = img.astype(np.float32)
    if img.max() > img.min():
        img = (img - img.min()) / (img.max() - img.min())
    return img


def run_preprocessing():
    """
    CHANGE: reads from CoND subfolder structure:
        data/raw/CoND/differentiated/*.png   → label 1
        data/raw/CoND/undifferentiated/*.png → label 0

    Returns image_manifest: one row per image (not per patch).
    """
    raw_dir  = PROJECT_ROOT / cfg['paths']['raw_dir']
    pre_dir  = PROJECT_ROOT / cfg['paths']['preprocessed_dir']
    pre_dir.mkdir(parents=True, exist_ok=True)

    target_h, target_w = cfg['preprocessing']['target_size']   # 200, 200

    class_map = {'differentiated': 1, 'undifferentiated': 0}   # folder → label
    manifest  = []

    for class_name, label in class_map.items():
        class_dir = raw_dir / class_name
        if not class_dir.exists():
            print(f"  WARNING: {class_dir} not found — skipping.")
            continue

        images = sorted(class_dir.glob('*.png'))
        print(f"  {class_name}: {len(images)} images  (label={label})")

        for img_path in images:
            img = load_image(img_path)

            # Resize (usually no-op — CoND is already 200×200)
            img = cv2.resize(img, (target_w, target_h), interpolation=cv2.INTER_LINEAR)

            # CLAHE — enhances contrast for phase contrast microscopy
            img_uint8 = (img * 255).astype(np.uint8)
            clahe     = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
            img_uint8 = clahe.apply(img_uint8)

            # Gaussian blur — reduces high-frequency noise
            img_uint8 = cv2.GaussianBlur(img_uint8, (3, 3), 0)

            save_name = f"{class_name}_{img_path.stem}_pre.png"
            cv2.imwrite(str(pre_dir / save_name), img_uint8)

            manifest.append({
                'image_id'    : img_path.stem,
                'preprocessed': save_name,
                'class_name'  : class_name,
                'label'       : label,
            })

    df = pd.DataFrame(manifest)
    print(f"\nTotal preprocessed : {len(df)}")
    print(f"Label distribution :\n{df['label'].value_counts().to_string()}")
    return df


image_manifest = run_preprocessing()
image_manifest.head()
```

## Module 2 — Patch Generation → **REMOVED**

**Why removed:**  
CoND images are already 200×200 px crops of *individual cells*. In the prototype, full microscopy field images (512×512) were divided into a 3×3 grid to create 9 patches per image — each patch covering a sub-region of the field.

CoND has already done this: each image *is* a single cell. Patching would split a single cell into 9 meaningless sub-regions, destroying the morphological features we need to measure.

**Result:** `image_manifest` (one row per image) directly replaces `patch_manifest`.

```python
# Module 2 intentionally not implemented.
print("Module 2 (Patching): SKIPPED")
print(f"Reason  : CoND images are already 200×200 individual-cell crops.")
print(f"Samples : {len(image_manifest)} (one per image, not per patch)")
```

## Module 3 — Feature Extraction

**What changed:**  
The extraction function (`extract_features`) is **identical** to the prototype — same 10 morphological features, same watershed segmentation. The only change is that it now receives a full 200×200 image (not a 170×170 patch).

The 10 features are:
- `cell_count`, `soma_area_mean`, `soma_area_total`, `perimeter_mean`
- `circularity_mean`, `eccentricity_mean`, `solidity_mean`
- `neurite_length`, `branch_count`, `neurite_density`

```python
def segment_image(img_float):
    """Background subtraction + watershed segmentation for phase contrast."""
    background = gaussian_filter(img_float, sigma=20)
    deviation  = np.abs(img_float - background)
    thresh     = np.percentile(deviation, 88)
    binary     = deviation > thresh
    binary     = morphology.remove_small_objects(binary, min_size=150)
    binary     = morphology.remove_small_holes(binary, area_threshold=300)

    distance   = ndi.distance_transform_edt(binary)
    coords     = peak_local_max(distance, min_distance=15, labels=binary)
    mask       = np.zeros(distance.shape, dtype=bool)
    mask[tuple(coords.T)] = True
    markers, _ = ndi.label(mask)
    labels     = watershed(-distance, markers, mask=binary)
    return labels > 0


def extract_features(image_path):
    """Extract 10 morphological features from a single preprocessed image."""
    img       = cv2.imread(str(image_path), cv2.IMREAD_GRAYSCALE)
    img_float = img.astype(np.float32) / 255.0
    binary    = segment_image(img_float)

    labeled  = measure.label(binary)
    regions  = measure.regionprops(labeled, intensity_image=img_float)
    if not regions:
        return None

    areas          = [r.area          for r in regions]
    perimeters     = [r.perimeter     for r in regions]
    circularities  = [4 * np.pi * r.area / (r.perimeter ** 2)
                      if r.perimeter > 0 else 0 for r in regions]
    eccentricities = [r.eccentricity  for r in regions]
    solidities     = [r.solidity      for r in regions]

    skeleton        = skeletonize(binary)
    neurite_length  = int(skeleton.sum())
    kernel          = np.ones((3, 3), dtype=int)
    neighbour_count = convolve(skeleton.astype(int), kernel)
    branch_points   = int(((neighbour_count >= 4) & skeleton).sum())
    total_area      = binary.sum()
    neurite_density = neurite_length / total_area if total_area > 0 else 0

    return {
        'cell_count'       : len(regions),
        'soma_area_mean'   : np.mean(areas),
        'soma_area_total'  : np.sum(areas),
        'perimeter_mean'   : np.mean(perimeters),
        'circularity_mean' : np.mean(circularities),
        'eccentricity_mean': np.mean(eccentricities),
        'solidity_mean'    : np.mean(solidities),
        'neurite_length'   : neurite_length,
        'branch_count'     : branch_points,
        'neurite_density'  : neurite_density,
    }


def run_feature_extraction():
    pre_dir      = PROJECT_ROOT / cfg['paths']['preprocessed_dir']
    features_dir = PROJECT_ROOT / cfg['paths']['features_dir']
    features_dir.mkdir(parents=True, exist_ok=True)

    feature_cols = [
        'cell_count', 'soma_area_mean', 'soma_area_total',
        'perimeter_mean', 'circularity_mean', 'eccentricity_mean',
        'solidity_mean', 'neurite_length', 'branch_count', 'neurite_density'
    ]
    rows = []

    for _, row in image_manifest.iterrows():
        img_path = pre_dir / row['preprocessed']
        feats    = extract_features(img_path)
        if feats is None:
            print(f"  WARNING: No regions in {row['preprocessed']} — skipping.")
            continue
        entry = {
            'image_id'  : row['image_id'],
            'class_name': row['class_name'],
            'label'     : row['label'],
        }
        entry.update(feats)
        rows.append(entry)

    df = pd.DataFrame(rows)
    df.to_csv(features_dir / 'raw_features.csv', index=False)
    print(f"Extracted : {len(df)} images")
    print(f"Columns   : {list(df.columns)}")
    print(f"\nLabel distribution:\n{df['label'].value_counts().to_string()}")
    return df, feature_cols


features_df, feature_cols = run_feature_extraction()
features_df.describe()
```

## Modules 4a / 4b — Split into Before & After Tables

**What changed:**  
The prototype split rows by `timepoint == 'before'` / `'after'` — a column that came from parsing a timestamp in the filename.

CoND images have no timestamps. Instead, the split is:
- `label == 0` → undifferentiated → "before" state
- `label == 1` → differentiated   → "after" state

This is the same conceptual split — one group represents the pre-treatment morphology, the other the post-treatment morphology — but driven by the CoND class label rather than a filename timestamp.

```python
# CHANGE: split by label (0/1), not by timepoint from filename
before_df = features_df[features_df['label'] == 0].copy().reset_index(drop=True)
after_df  = features_df[features_df['label'] == 1].copy().reset_index(drop=True)

print(f"Before (undifferentiated) : {len(before_df)} images")
print(f"After  (differentiated)   : {len(after_df)}  images")
```

## Module 5 — Delta Computation (Synthetic Cross-Sectional Pairing)

**What changed and why:**

| | Prototype | CoND |
|---|---|---|
| **Pairing method** | ID-matched — same `base_id` + same patch position | Random pairing — fixed seed 42 |
| **Biological meaning** | Longitudinal within-sample | Cross-sectional population-level |
| **Why this works** | Same cell → true before/after | Undiff population ≈ "before state", Diff population ≈ "after state" |

**Academic positioning:**  
This is documented as `pair_type = 'synthetic_cross_sectional'`. The delta features represent the *expected population-level morphological shift* from undifferentiated to differentiated, not within-sample change. This is an approximation. When the real Ag-CS NP dataset arrives (with true before/after pairs from the same sample), Module 5 can be replaced with true ID matching and the rest of the pipeline is unchanged.

**Why this is better than mixing LIVECell + CoND:**  
Using one consistent dataset (CoND, same cell type, same imaging modality) with cross-sectional approximation is more coherent than computing deltas from glioblastoma cells (LIVECell A172) and classifying with neuroblastoma cells (CoND).

```python
def run_delta_computation():
    """
    CHANGE: randomly pair each undifferentiated image with one differentiated
    image (fixed seed = 42) instead of matching by ID.
    delta = after_feature - before_feature for each of the 10 features.
    """
    rng = np.random.default_rng(cfg['pairing']['random_seed'])

    # Shuffle after-pool once (fixed seed → reproducible)
    after_shuffled = after_df.sample(
        frac=1, random_state=cfg['pairing']['random_seed']
    ).reset_index(drop=True)

    n_after    = len(after_shuffled)
    delta_rows = []

    for i, b_row in before_df.iterrows():
        a_row = after_shuffled.iloc[i % n_after]   # cycle if more before than after

        entry = {
            'before_image_id': b_row['image_id'],
            'after_image_id' : a_row['image_id'],
            'label'          : 1,                  # pair = one differentiation event
            'pair_type'      : cfg['pairing']['pair_type'],
        }
        for col in feature_cols:
            entry[f'delta_{col}'] = float(a_row[col]) - float(b_row[col])

        delta_rows.append(entry)

    df = pd.DataFrame(delta_rows)
    print(f"Delta rows  : {len(df)}")
    print(f"Pair type   : {df['pair_type'].iloc[0]}")
    print(f"\nSample (first 3 rows):")
    print(df[['before_image_id', 'after_image_id',
              'delta_neurite_length', 'delta_branch_count']].head(3).to_string(index=False))
    return df


delta_df = run_delta_computation()
delta_df.head()
```

## Module 6 — Tabular Dataset Construction

**What changed:**

- `label` column is now **real** (binary 0/1 from CoND folder). In the prototype it was `None` because there were no ground-truth labels, and `delta_neurite_length` was used as a proxy.
- **No `concentration` column** — CoND has no concentration metadata (binary labels only).
- `pair_type` column is added to document that pairs are synthetic.
- Everything else (before/after/delta prefixed columns, merge logic) is identical.

```python
def run_dataset_construction():
    features_dir = PROJECT_ROOT / cfg['paths']['features_dir']

    # Prefix before/after feature columns for clarity
    before_feat = before_df[['image_id'] + feature_cols].copy()
    before_feat = before_feat.rename(columns={
        col: f'before_{col}' for col in feature_cols
    }).rename(columns={'image_id': 'before_image_id'})

    after_feat = after_df[['image_id'] + feature_cols].copy()
    after_feat = after_feat.rename(columns={
        col: f'after_{col}' for col in feature_cols
    }).rename(columns={'image_id': 'after_image_id'})

    dataset = delta_df.merge(before_feat, on='before_image_id', how='left')
    dataset = dataset.merge(after_feat,  on='after_image_id',  how='left')
    dataset['is_synthetic'] = False

    # Ordered column layout
    id_cols     = ['before_image_id', 'after_image_id', 'pair_type']
    before_cols = [f'before_{col}' for col in feature_cols]
    after_cols  = [f'after_{col}'  for col in feature_cols]
    delta_cols_ = [f'delta_{col}'  for col in feature_cols]
    meta_cols   = ['label', 'is_synthetic']

    dataset = dataset[id_cols + before_cols + after_cols + delta_cols_ + meta_cols]

    out_path = features_dir / 'dataset.csv'
    dataset.to_csv(out_path, index=False)

    print(f"Dataset shape   : {dataset.shape}")
    print(f"Label breakdown :\n{dataset['label'].value_counts().to_string()}")
    print(f"\nColumn groups:")
    print(f"  Before  : {len(before_cols)} features  (e.g. before_soma_area_mean)")
    print(f"  After   : {len(after_cols)} features  (e.g. after_soma_area_mean)")
    print(f"  Delta   : {len(delta_cols_)} features  (e.g. delta_neurite_length)")
    print(f"\nSaved to: {out_path}")
    return dataset, before_cols, after_cols, delta_cols_


dataset, before_cols, after_cols, delta_cols = run_dataset_construction()
dataset.head()
```

## Module 7 — Tabular Jitter Augmentation

**No changes.** The augmentation logic is identical to the prototype — sample a real row, add Gaussian noise scaled by `alpha × std(feature)`. The column names are inherited from Module 6, so no code changes are needed here.

```python
def run_augmentation(dataset, before_cols, after_cols, delta_cols):
    features_dir = PROJECT_ROOT / cfg['paths']['features_dir']

    alpha       = cfg['augmentation']['alpha']
    target_rows = cfg['augmentation']['target_rows']
    rng         = np.random.default_rng(cfg['augmentation']['random_seed'])
    aug_cols    = before_cols + after_cols + delta_cols

    real_df  = dataset[dataset['is_synthetic'] == False].copy()
    col_stds = real_df[aug_cols].std()

    rows_needed    = target_rows - len(real_df)
    synthetic_rows = []

    while len(synthetic_rows) < rows_needed:
        base = real_df.sample(1, random_state=int(rng.integers(1e6))).iloc[0].copy()
        for col in aug_cols:
            base[col] += rng.normal(0, alpha * col_stds[col])
        base['is_synthetic'] = True
        synthetic_rows.append(base)

    synthetic_df = pd.DataFrame(synthetic_rows)
    augmented_df = pd.concat([real_df, synthetic_df], ignore_index=True)

    out_path = features_dir / 'augmented_dataset.csv'
    augmented_df.to_csv(out_path, index=False)

    print(f"Real rows      : {len(real_df)}")
    print(f"Synthetic rows : {len(synthetic_df)}")
    print(f"Total rows     : {len(augmented_df)}")
    return augmented_df


augmented_df = run_augmentation(dataset, before_cols, after_cols, delta_cols)
print(f"Augmented shape: {augmented_df.shape}")
```

## Module 8 — ML Model Training

**What changed:**

| | Prototype | CoND |
|---|---|---|
| **XGBoost** | `XGBRegressor` | `XGBClassifier` |
| **Random Forest** | `RandomForestRegressor` | `RandomForestClassifier` |
| **SVM** | `SVR` (regression) | `SVC` (classification, `probability=True` for ROC-AUC) |
| **Target variable** | `delta_neurite_length` (proxy — no real labels) | `label` (real 0/1 from CoND folders) |
| **Metrics** | RMSE, MAE, R² | Accuracy, F1, ROC-AUC |

Everything else — train/test split, StandardScaler for SVC, test-set composition (real rows only), training on real+synthetic — is unchanged.

```python
def run_ml_training(augmented_df, before_cols, after_cols, delta_cols):
    models_dir  = PROJECT_ROOT / cfg['paths']['models_dir']
    results_dir = PROJECT_ROOT / cfg['paths']['results_dir']
    models_dir.mkdir(parents=True, exist_ok=True)
    results_dir.mkdir(parents=True, exist_ok=True)

    X_cols = before_cols + after_cols + delta_cols

    # Test set = real rows only (no augmented data)
    real_data  = augmented_df[augmented_df['is_synthetic'] == False].copy()
    synth_data = augmented_df[augmented_df['is_synthetic'] == True].copy()

    n_test   = max(1, int(len(real_data) * cfg['ml']['test_size']))
    test_df  = real_data.sample(n=n_test, random_state=SEED)
    train_df = pd.concat([real_data.drop(test_df.index), synth_data],
                         ignore_index=True)

    X_train, y_train = train_df[X_cols].fillna(0).values, train_df['label'].values.astype(int)
    X_test,  y_test  = test_df[X_cols].fillna(0).values,  test_df['label'].values.astype(int)

    scaler     = StandardScaler()
    X_train_sc = scaler.fit_transform(X_train)
    X_test_sc  = scaler.transform(X_test)

    print(f"Train rows : {len(X_train)} (real: {len(real_data) - n_test}, synthetic: {len(synth_data)})")
    print(f"Test rows  : {len(X_test)} (real only)")
    print(f"Features   : {len(X_cols)}")
    print(f"Class balance (test): {dict(pd.Series(y_test).value_counts())}\n")

    # ── CHANGE: classification models ──────────────────────────────────────────
    models = {
        'xgboost'       : xgb.XGBClassifier(n_estimators=100, random_state=SEED,
                              verbosity=0, eval_metric='logloss'),
        'random_forest' : RandomForestClassifier(n_estimators=100, random_state=SEED),
        'svc'           : SVC(kernel='rbf', C=1.0, probability=True, random_state=SEED),
    }

    results = {}
    for name, model in models.items():
        Xtr = X_train_sc if name == 'svc' else X_train
        Xte = X_test_sc  if name == 'svc' else X_test

        model.fit(Xtr, y_train)
        preds      = model.predict(Xte)
        preds_prob = model.predict_proba(Xte)[:, 1]

        # ── CHANGE: classification metrics ─────────────────────────────────────
        acc     = accuracy_score(y_test, preds)
        f1      = f1_score(y_test, preds, zero_division=0)
        roc_auc = roc_auc_score(y_test, preds_prob)

        results[name] = {'accuracy': acc, 'f1': f1, 'roc_auc': roc_auc, 'model': model}

        print(f"── {name.upper()} ──")
        print(f"  Accuracy : {acc:.4f}")
        print(f"  F1       : {f1:.4f}")
        print(f"  ROC-AUC  : {roc_auc:.4f}")
        print(classification_report(y_test, preds,
              target_names=['undifferentiated', 'differentiated'], zero_division=0))

    best = max(results, key=lambda k: results[k]['roc_auc'])
    print(f"\nBest model (ROC-AUC): {best}  ({results[best]['roc_auc']:.4f})")
    return results, X_cols, X_test, X_test_sc, y_test, scaler


results, X_cols, X_test, X_test_sc, y_test, scaler = run_ml_training(
    augmented_df, before_cols, after_cols, delta_cols
)
```

## Module 9 — Interpretability (SHAP)

**No structural changes.** `TreeExplainer` works identically for `XGBClassifier` and `RandomForestClassifier`. The only adjustment is that `shap_values` for classifiers may return a list (one array per class) — we index `[1]` to get SHAP values for the positive class (differentiated).

Permutation importance for SVC is unchanged; scoring is `roc_auc` (was also the most appropriate metric in the prototype).

**What this answers for your capstone:**  
"Which morphological features — measured across the before state, the after state, and the delta — are most important for distinguishing differentiated from undifferentiated SH-SY5Y cells?" 

```python
def run_interpretability(results, X_cols, X_test, X_test_sc, y_test):
    plots_dir = PROJECT_ROOT / cfg['paths']['plots_dir']
    plots_dir.mkdir(parents=True, exist_ok=True)
    top_n = cfg['interpretability']['top_n_features']

    # ── SHAP — XGBoost & Random Forest ───────────────────────────────────────
    for name in cfg['interpretability']['shap_models']:
        model     = results[name]['model']
        explainer = shap.TreeExplainer(model)
        shap_vals = explainer.shap_values(X_test)

        # Classifiers may return list [class_0, class_1] → take class 1
        if isinstance(shap_vals, list):
            shap_vals = shap_vals[1]

        mean_abs = np.abs(shap_vals).mean(axis=0)
        shap_df  = pd.DataFrame({
            'feature': X_cols, 'importance': mean_abs
        }).sort_values('importance', ascending=False).head(top_n)

        print(f"\n── SHAP: {name.upper()} — Top {top_n} features ──")
        print(shap_df.to_string(index=False))

        fig, ax = plt.subplots(figsize=(9, 4))
        ax.barh(shap_df['feature'][::-1], shap_df['importance'][::-1])
        ax.set_xlabel('Mean |SHAP value|  →  contribution to differentiation prediction')
        ax.set_title(
            f'SHAP Feature Importance — {name.upper()}\n'
            f'Target: differentiated vs undifferentiated  |  Dataset: CoND'
        )
        plt.tight_layout()
        plt.savefig(plots_dir / f'shap_{name}_CoND.png', dpi=150)
        plt.show()

    # ── Permutation Importance — SVC ─────────────────────────────────────────
    print(f"\n── Permutation Importance: SVC — Top {top_n} features ──")
    perm = permutation_importance(
        results['svc']['model'],
        X_test_sc, y_test,
        n_repeats=cfg['interpretability']['n_repeats'],
        random_state=SEED,
        scoring='roc_auc'
    )
    perm_df = pd.DataFrame({
        'feature': X_cols, 'importance': perm.importances_mean
    }).sort_values('importance', ascending=False).head(top_n)

    print(perm_df.to_string(index=False))

    fig, ax = plt.subplots(figsize=(9, 4))
    ax.barh(perm_df['feature'][::-1], perm_df['importance'][::-1])
    ax.set_xlabel('Mean decrease in ROC-AUC when feature is permuted')
    ax.set_title(
        'Permutation Importance — SVC\n'
        'Target: differentiated vs undifferentiated  |  Dataset: CoND'
    )
    plt.tight_layout()
    plt.savefig(plots_dir / 'permutation_svc_CoND.png', dpi=150)
    plt.show()

    print(f"\nAll plots saved to: {plots_dir}")


run_interpretability(results, X_cols, X_test, X_test_sc, y_test)

print("\n" + "="*60)
print("Pipeline complete.")
print("="*60)
```
